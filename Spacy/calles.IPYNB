```python
import osmnx as ox
import pandas as pd

# Configurar el nombre de la ciudad
ciudad = "Monterrey, Nuevo León, Mexico"

# Descargar el grafo de calles de la ciudad desde OpenStreetMap
grafo = ox.graph_from_place(ciudad, network_type="all")

# Obtener las calles y sus atributos
calles = ox.graph_to_gdfs(grafo, nodes=False, edges=True)

# Filtrar para obtener únicamente las columnas relevantes
calles_df = calles[['name', 'highway', 'osmid']].dropna(subset=['name'])

# Crear un DataFrame con nombres de calles únicos
calles_unicas = calles_df[['name']].drop_duplicates().reset_index(drop=True)

# Añadir la columna de colonias (barrios)
barrios = ox.geocode_to_gdf(ciudad)
calles_unicas['colonia'] = ox.geocode_to_gdf(ciudad)['display_name'][0]

# Guardar el dataset en un archivo CSV
calles_unicas.to_csv('calles_colonias_monterrey.csv', index=False)

print("Dataset generado y guardado como 'calles_colonias_monterrey.csv'")
```
